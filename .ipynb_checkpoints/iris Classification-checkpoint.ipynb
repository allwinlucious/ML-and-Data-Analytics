```python
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, KFold
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix , classification_report, accuracy_score
import numpy as np
```

```python
iris_dataset = datasets.load_iris()
```

```python
 iris_dataset.target
```

```python
iris_dataset.target_names
```

```python
iris_dataset.data
```

```python
df = pd.DataFrame(iris_dataset.data ,columns=iris_dataset.feature_names)
df
```

```python
df['target'] =  iris_dataset.target
```

```python
df
```

```python
for c in df.columns:
    df[c].plot(title=c)
    plt.show()
```

```python
df.plot(kind = "density", subplots = True,layout =(5,2),sharex = False,figsize=(20,20))
plt.show()
```

```python
X = df[df.columns[0:-1]]
y = df[df.columns[-1]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
clf = DecisionTreeClassifier(random_state=0)
```

```python
clf.fit(X_train,y_train)
```

```python
y_predicted = clf.predict(X_test)
y_predicted
```

```python
confusion_matrix(y_test,y_predicted)
```

```python
classification_report(y_test,y_predicted)
```

   ### 3 fold cross validation
    

```python
kf = KFold(n_splits=3,shuffle = True)
KNN = KNeighborsClassifier(n_neighbors=3)
LG = LogisticRegression(max_iter=3000)
```

```python
KNN_scores = LG_scores = clf_scores = []
```

```python
for train_index, test_index in kf.split(X):
    X_train, X_test = X.loc[train_index], X.loc[test_index]
    y_train, y_test = y.loc[train_index], y.loc[test_index]
    KNN.fit(X_train,y_train)
    LG.fit(X_train,y_train)
    clf.fit(X_train,y_train)
    KNN_pred = KNN.predict(X_test)
    LG_pred = LG.predict(X_test)
    clf_pred = clf.predict(X_test)
    KNN_accu = accuracy_score(y_test,KNN_pred)
    KNN_scores = np.append(KNN_scores,KNN_accu)
    LG_accu = accuracy_score(y_test,LG_pred)
    LG_scores = np.append(LG_scores,LG_accu)
    clf_accu = accuracy_score(y_test,clf_pred)
    clf_scores = np.append(clf_scores,clf_accu)
    
```

```python
np.mean(KNN_scores)
```

```python
np.mean(LG_scores)
```

```python
np.mean(clf_scores)
```
