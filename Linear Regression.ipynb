```python
import pandas as pd
import matplotlib.pyplot as plt
from  sklearn.model_selection import train_test_split 
from  sklearn.linear_model import LinearRegression
import sklearn.metrics
```

```python
data_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/concrete/compressive/Concrete_Data.xls"
df = pd.read_excel(data_url)
```

```python
df
```

```python
df.columns
```

```python
for c in df.columns:
    df[c].plot(title = c)
    plt.show()
```

```python
df.plot(kind = "density", subplots = True,layout =(5,2),sharex = False,figsize=(20,20))
plt.show()
```

```python
X = df[df.columns[0:-1]] #excluding last column
y = df[df.columns[-1]]#last column
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
reg = LinearRegression().fit(X_train,y_train)
```

```python
reg.score(X_train,y_train)
```

```python
y_predicted = reg.predict(X_test)
y_predicted
```

## evaluate the predictions

```python
mean_square_error = sklearn.metrics.mean_squared_error(y_test,y_predicted)
print(mean_square_error)
```

```python
r2_score = sklearn.metrics.r2_score(y_test,y_predicted)
r2_score
```
